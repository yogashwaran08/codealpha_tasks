# Importing Necessary Libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# Load dataset

```python
df = pd.read_csv('student_habits_performance.csv')
```

# Set style


```python
sns.set(style='whitegrid')
```

# 1. Distribution Plots

```python
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
sns.histplot(df['study_hours_per_day'], bins=20, kde=True, color='skyblue')
plt.title('Study Hours Per Day Distribution')

plt.subplot(1, 2, 2)
sns.histplot(df['exam_score'], bins=20, kde=True, color='salmon')
plt.title('Exam Score Distribution')
plt.tight_layout()
plt.show()
```

# 2. Correlation Heatmap

```python
plt.figure(figsize=(10, 8))
numeric_cols = df.select_dtypes(include=['float64', 'int64']).columns
sns.heatmap(df[numeric_cols].corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap')
plt.show()
```

# 3. Scatter Plot: Study Hours vs Exam Score

```python
plt.figure(figsize=(7, 5))
sns.scatterplot(x='study_hours_per_day', y='exam_score', hue='gender', data=df)
plt.title('Study Hours vs Exam Score')
plt.show()
```

# 4. Boxplot: Exam Score by Part-Time Job

```python
plt.figure(figsize=(6, 5))
sns.boxplot(x='part_time_job', y='exam_score', data=df)
plt.title('Exam Score by Part-Time Job')
plt.show()
```

# 5. Barplot: Average Exam Score by Diet Quality

```python
plt.figure(figsize=(7, 5))
sns.barplot(x='diet_quality', y='exam_score', data=df, estimator='mean', ci='sd')
plt.title('Average Exam Score by Diet Quality')
plt.show()
```

# 6. Pairplot: Multivariate relationships

```python
sns.pairplot(df[['study_hours_per_day', 'sleep_hours', 'exercise_frequency', 'exam_score']], kind='scatter')
plt.suptitle('Pairwise Relationships', y=1.02)
plt.show()
```
